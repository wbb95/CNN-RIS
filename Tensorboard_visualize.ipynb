```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
FER2013_train_acc=event_accumulator.EventAccumulator(r'models/FER2013_AntCNN_71.998/epoch/accuracy/train/events.out.tfevents.1594204258.scnu')
FER2013_train_acc.Reload()
print(FER2013_train_acc.scalars.Keys())
FER2013_valid_acc=event_accumulator.EventAccumulator(r'models/FER2013_AntCNN_71.998/epoch/accuracy/valid/events.out.tfevents.1594204258.scnu')
FER2013_valid_acc.Reload()
print(FER2013_valid_acc.scalars.Keys())
```

```python
train_acc=FER2013_train_acc.scalars.Items('epoch/accuracy')
valid_acc=FER2013_valid_acc.scalars.Items('epoch/accuracy')
#print(len(train_acc))
#print([(i.step,i.value) for i in train_acc])
#print(len(valid_acc))
#print([(i.step,i.value) for i in valid_acc])
fig=plt.figure(figsize=(6,4))
ax1=fig.add_subplot(111)

ax1.plot([i.step for i in train_acc],[i.value for i in train_acc],label='train_acc')
ax1.set_xlim(0)

ax1.plot([i.step for i in valid_acc],[i.value for i in valid_acc],label='valid_acc')
ax1.set_xlabel("epoch")
ax1.set_ylabel("")

plt.title('FER2013 basic Data Augmentation')
plt.legend(loc='lower right')
plt.savefig("FER2013 basic Augmentation.png")
plt.show()
```
